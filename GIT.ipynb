### Stappenplan voor git bij gitlab

In dit document volg je een korte demo van het gebruik van git voor het gitlab platform. Je hebt een account bij gitlab nodig om dit te volgen.

We gaan een repository maken en daar twee bestanden in zetten om kennis te maken met de manier waarop git werkt en je kan ondersteunen bij programmeerwerkzaamheden.

Om dit te laten werken heb je git nodig op je computer. Dit doe je door naar de volgende website te gaan en het installatiepakket te downloaden en installeren.

https://git-scm.com/downloads

Let op dat je ook GUI's (grafische interface) kunt downloaden die het werk makkelijker maken maar voor een goede introductie doen wij dit via de command line.

#### gitlab - start nieuw project

* Ga naar gitlab en zorg dat je ingelogd bent.
* Maak een nieuw project aan via de 'New Project' knop en geef het de naam 'testproject'

![title](img/git01.PNG)

* Selecteer Create Blank Project
* Vul als Project name 'testproject' in en druk op Create Project

![title](img/git02.PNG)

Als alles goed is gegaan zie je de melding 'Project 'testproject' was successfully created.' en heb je je eerste repository aangemaakt! Houd de laatste pagina met alle details op gitlab open voor de volgende actie.

#### git clone

* Start de command line en ga ouderwets met het commando cd naar het juiste pad waar je je code wilt downloaden.
* Om de repo te downloaden tik je ```git clone git@gitlab.com:<jouw gitlabnaam>/testproject.git``` in
* Voila, de code staat nu netjes in een submap met de naam testproject

#### git status en git add

* Ga met explorer naar de map een maak in de map testproject een bestand aan genaamd 'bestand_1.txt'
* Schrijf een stukje tekst in het bestand en sla het op
* Ga naar de commandline terug en tik ```git status``` in

![title](img/git03.PNG)

Je ziet dat git ontdekt heeft dat er een bestand is toegevoegd wat niet gevolgd wordt. Om dit aan de repo toe te voegen hebben we git add nodig.

* Run het commando ```git add bestand_1.txt```
* Run nog een keer het commando ```git status```

![title](img/git04.PNG)

Je ziet nu dat het bestand als new file wordt gezien. Vanaf nu worden de wijzigingen in dit bestand door git bijgehouden

#### git commit en git push

Om de veranderingen ook op gitlab door te voeren is het commando ```git push``` nodig maar voordat je dat doet moet je eerst vertellen wat je gedaan hebt via een ```git commit```

* Run het commando ```git commit -am"eerste upload"```

Dit zorgt ervoor dat in het logboek een aantekening komt met jouw opmerking waardoor je snel kunt zien wat je precies bij deze wijziging gedaan hebt. In dit geval wil je een eerste upload doen van je nieuwe bestand. Om de wijzigingen nu naar gitlab door te voeren moet je het volgende commando invoeren;

* Run het commando ```git push```

Let op dat je soms eerste de volgende commando's moet uitvoeren (alleen bij de allereerste git actie). Dit wordt overigens ook vermeld op de gitlab pagina na het aanmaken van je repo.

* ```git config --global user.name "je naam"```
* ```git config --global user.email "je email"```

Het push commando synct de code op je computer met de gitlab server. 

Je kunt nu op je gitlab pagina naar Project Overview gaan en daar je bestand zien staan.

![title](img/git05.PNG)

#### git pull

Om te simuleren dat iemand anders ook aan je project werkt klik je nu op de website op de bestandsnaam en tik je edit in.

![title](img/git06.PNG)

* Voeg nog een extra regel aan je bestand toe
* Vul onder Commit message 'aanpassing via IDE' in
* Druk op 'Commit changes'

![title](img/git07.PNG)

Als alles goed is gegaan zie je het bericht 'Your changes have been successfully committed.' 

Ga terug naar je eigen map en bekijk de inhoud van bestand_1.txt. Er is nog niks veranderd. Dat komt omdat je nu moet zorgen dat de inhoud van gitlab naar jouw pc gesynct wordt. Dit doe je met het volgende commando in de command line;

* ```git pull```

Door het uitvoeren van een 'pull' synchroniseer je de bestanden op gitlab met je lokale pc. Je ziet in het kort ook een samenvatting van wat git tegen is gekomen;

![title](img/git08.PNG)

Bekijk je bestand nog een keer en je zult zien dat je de extra regel nu ook in je lokale bestand hebt.

#### git checkout

Stel nu dat je het helemaal niet eens bent met de wijziging en terug wilt naar de eerdere versie. In de echte wereld zou het bijvoorbeeld kunnen zijn dat je helemaal vastgelopen bent met je code. Geen probleem voor git want je kunt altijd terug naar elke commit die je gemaakt hebt. 

* ga in de browser naar gitlab en het menu item Repository | Commits

Je ziet hier een overzicht van alle commits die je gedaan hebt. 

![title](img/git09.PNG)

Helemaal rechts zie je de zogenaamde SHA, in mijn geval is dat b6fef600, dit is een unieke id die hoort bij de commit en bestaat eigenlijk uit nog meer hexadecimale getallen. (Let op dat je zowel de korte code als de lange code als SHA kunt gebruiken. We gebruiken nu de lange code omdat gitlab dat makkelijk kopieert) Klik op het kopieer icoontje rechts naar de SHA. Dit kopieert de gehele SHA in het geheugen. Ga naar de commandline en tik in;

* ```git checkout <jouw SHA>```

Je kunt de SHA heel eenvoudig kopieren door in de commandline op de rechtermuisknop te drukken, dit plakt de waarde die je net gekopieert hebt in de commandline waardoor je niet al die hex getallen in hoeft te voeren.

Bekijk nu weer je tekstbestand. De wijziging is ongedaan gemaakt.

Zo kun je altijd bij je oude code komen!

#### Geavanceerder

Git heeft nog veel meer mogelijkheden zoals het maken van aparte splitsingen in je code (branches) wat handig kan zijn als je met meer mensen aan een project werkt en ieder een eigen deel doet. Daarnaast kun je branches weer terug brengen naar de hoofdtak via pull requests waar je ook de kans krijgt om de code te reviewen etc. etc. Git is een studie op zich.
