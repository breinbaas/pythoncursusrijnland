### Bepaal noodzakelijke onderdelen en mogelijke knelpunten

Als je je projectdoel kent is het een goed idee om na te gaan waar je wellicht tegenaan zult gaan lopen. Je kunt je dan voorbereiden op die moeilijke onderdelen en nagaan of het haalbaar voor je is. Hier een praktisch (maar heerlijk onrealistisch) voorbeeld.

#### Beleggingsapp

![beleggingsapp](img/rijk.jpg)

**Idee van een ficitieve ik**

`Ik wil een app bouwen die voor mij de beurskoers gaat voorspellen en waarmee ik automatisch kan beleggen en miljonair word. Ik wil dat vanaf mijn telefoon bedienen en berichten via Whatsapp krijgen.`

**Noodzakelijke onderdelen - brainstorm**

* ik moet ergens beurskoersen kunnen downloaden, het liefst zo actueel mogelijk
* ik moet aandelen kunnen kopen en verkopen
* ik moet de beurskoersen gaan voorspellen
* ik moet een user interface voor een telefoon maken
* ik moet met whatsapp kunnen communiceren

**De volgende stap is dan nagaan wat je kunt, bijvoorbeeld;**

* Koersen van aandelen downloaden is geen probleem want ik weet hoe ik met API's om moet gaan en ik weet dat er websites zijn die koersinformatie (al dan niet betaald) leveren. 
* Beurskoersen voorspellen denk ik te kunnen met een tijdreeks analyse

**Daarna is het zinvol om eens te gaan rondkijken met je favo zoekmachine.. denk aan de volgende zaken;**

* is er al een app die dit kan? zonde van de tijd als ik iets opnieuw bouw (maar soms ook wel weer leuk ;-) )
* zijn er bibliotheken die mij kunnen helpen?
* zijn mijn doelen voor de onderdelen die ik nog niet weet realistisch?
* kan ik een go - no go lijstje maken om te zien of sommige onderdelen wellicht niet kunnen en daarmee het hele proces stil leggen

**In het geval van de app zou je bijvoorbeeld tot de volgende realisatie kunnen komen.**

* er is geen app die mij automatisch miljonair maakt maar er zijn wel beleggingsapps
* het maken van een user interface in Python en voor een telefoon vereist dat ik een webinterface moet bouwen maar dat heb ik nog nooit gedaan, ik moet dus HTML, CSS en Javascript gaan leren
* ik vind nergens informatie dat ik Whatsapp via Python kan bedienen maar het schijnt dat het wel met Telegram kan
* ik lees overal dat het onmogelijk is om beurskoersen met tijdseries te voorspellen.. misschien is het toch niet zo makkelijk

**En wellicht kom je dan tot een hele andere conclusie;**

* ok, miljonair gaat t niet worden
* ik vind het wel leuk om de beurskoersen op te vragen en weet hoe dat moet
* dat idee om via Telegram berichten te krijgen is interessant
* ik ga toch eens kijken hoe mijn voorspelling presteert tegenover de werkelijke koersen

**En bouw je vervolgens een aangepaste app;**

Een app die beurskoersen kan downloaden en waarbij de gebruiker een bepaald aandeel aan kan geven waarna de app de koers om de 2 uur controleert en via Telegram een bericht geeft als de koers onder bedrag x gezakt is danwel boven bedrag y gestegen is. Daarnaast ga ik voor de gein eens die gedownloade koersen bewaren en kijken of ik in een Jupyter notebook een machine learning model kan bouwen.. als dat werkt kan ik dat weer in de app gaan inbouwen voor een volgende versie...

#### Bottomline

Denk groot, start klein en zorg ervoor dat je enthousiast blijft over je project door haalbare doelen te stellen. Naarmate je langer programmeert is het ook makkelijker om die haalbaarheid in te schatten. In het begin is dit gewoon moeilijker maar bedenk je dat er heel veel kan met Python!
